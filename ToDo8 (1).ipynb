#<Font size= 70 color = yellow> Terceira análise:

🧠 Dos funcionários que têm filhos, qual o número mais comum?
> Para essa tarefa foi necessário usar o método "query" do pandas para que retornasse uma tabela onde só aparecem os funcionários que possuem um valor de filhos maior que 0 e assim poderemos aplicar a função "mode" do pandas para que seja obtida a moda da quatidade de filhos dos funcionários.

```python
pd.DataFrame({'Moda Filhos':[funcionarios.query('Filhos >0')['Filhos'].mode().loc[0]]},index= ['Valor'])
```

#<Font size= 70 color = yellow> Quarta análise:

🧠 Qual a média e a mediana dos salários dos funcionários?
> Nessa atividade foi usada a função de média e mediana do Pandas "mean" e "median" respectivamente e depois foram atribuidos esse valores à um dicionário para que possa ser criado um dataframe com coluna e índices só para que fique mais bonitinho e visível.

O que esses valores podem indicar?

> Podemos analisar que os valores de média e mediana são bem próximos, logo, indica que não existem valores discrepantes (outliers) na coluna "Salario" o que significa que os valores estão assimetricos em sua distribuição.


```python
medMean= pd.DataFrame({'Valor':[funcionarios['Salario'].mean().round(),funcionarios['Salario'].median()]},index = ['Media','Mediana'])
medMean
```

#<Font size= 70 color = yellow> Quinta análise:

#<Font size= 70 color = yellow> Sexta análise:
🧠 Qual a média de salário e idade das pessoas casadas?
> Feito somente filtrando para que pegue as colunas "Salario" e "Anos" em que os valores da coluna "Est.civil" seja "casado" e assim retorne a média das duas colunas pedidas arredondando os valores e criando um DataFrame para melhor visualização dos dados obtidos.

```python
medias = pd.DataFrame(funcionarios[funcionarios['Est.civil'] == 'casado'][['Salario', 'Anos']].mean().round(2), columns = ['Media'])
medias
```

#<Font size= 70 color = yellow> Sétima análise:

🧠 Qual o funcionário que possui maior salário (informe seus atributos)?
> Aqui foi utilizado o método "loc" para localizar a linha com todos os atributos do funcionário que será retornado com a função "idxmax" para que a occorrencia máxima da coluna passada apareça e assim caia no "loc"

```python
maiorSalario = funcionarios.loc[[funcionarios['Salario'].idxmax()]]
maiorSalario
```

#<Font size= 70 color = yellow> Oitava análise:
🧠 Qual o funcionário que possui menor idade (informe seus atributos)?
> Na análise anterior utilizamos o "idxmax" e agora foi a vez do "idxmin" para que retorne a primeira ocorrência do valor mínimo na coluna passada e assim retorne todos os atributos atravez do "loc".

```python
menorIdade = funcionarios.loc[[funcionarios['Anos'].idxmin()]]
menorIdade
```
