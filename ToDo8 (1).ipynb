#<Font size= 70 color = yellow> Terceira an√°lise:

üß† Qual a m√©dia e a mediana dos sal√°rios dos funcion√°rios?
> Nessa atividade foi usada a fun√ß√£o de m√©dia e mediana do Pandas "mean" e "median" respectivamente e depois foram atribuidos esse valores √† um dicion√°rio para que possa ser criado um dataframe com coluna e √≠ndices s√≥ para que fique mais bonitinho e vis√≠vel.

#<Font size= 70 color = yellow> Sexta an√°lise:

#<Font size= 70 color = yellow> S√©tima an√°lise:

üß† Qual o funcion√°rio que possui maior sal√°rio (informe seus atributos)?
> Aqui foi utilizado o m√©todo "loc" para localizar a linha com todos os atributos do funcion√°rio que ser√° retornado com a fun√ß√£o "idxmax" para que a occorrencia m√°xima da coluna passada apare√ßa e assim caia no "loc"

```python
maiorSalario = funcionarios.loc[[funcionarios['Salario'].idxmax()]]
maiorSalario
```

#<Font size= 70 color = yellow> Oitava an√°lise:
üß† Qual o funcion√°rio que possui menor idade (informe seus atributos)?
> Na an√°lise anterior utilizamos o "idxmax" e agora foi a vez do "idxmin" para que retorne a primeira ocorr√™ncia do valor m√≠nimo na coluna passada e assim retorne todos os atributos atravez do "loc".

```python
menorIdade = funcionarios.loc[[funcionarios['Anos'].idxmin()]]
menorIdade
```
